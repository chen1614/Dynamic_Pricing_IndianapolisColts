```python
import pandas as pd
import numpy as np
secondary = pd.read_csv("secondary.csv")
```

```python
secondary
```

```python
ticket_price = {}
```

```python
for i in range(secondary.shape[0]):
    event = secondary["event_name"][i]
    section = secondary["section_name"][i]
    row = secondary["row_name"][i]
    seat = secondary["seat_num"][i]
    num = secondary["num_seats"][i]
    if secondary["activity_name"][i] in ("TE Posting","TE Price Update"):
        for j in np.arange(seat,seat+num):
#            if (event,section,seat,j) in ticket_price.keys():
#                ticket_price[(event,section,seat,j)] = secondary["te_posting_price"][i]
#            else:
            ticket_price[(event,section,seat,j)] = (secondary["event_date"][i],secondary["Orig_purchase_price"][i],secondary["add_datetime"][i],secondary["te_posting_price"][i],secondary["sales_channel"][i],secondary["season_year"][i],0)
    if secondary["activity_name"][i] == "TE Resale":
        for j in np.arange(seat,seat+num):
            ticket_price[(event,section,seat,j)] = (secondary["event_date"][i],secondary["Orig_purchase_price"][i],secondary["add_datetime"][i],secondary["te_purchase_price"][i]/num,secondary["sales_channel"][i],secondary["season_year"][i],1)
```

```python
secondary_new = pd.DataFrame(columns =['event_name', 'section_name','row_name','seat_num','event_date','Orig_purchase_price','add_datetime','final_price','sales_channel','season_year','target'])
```

```python
for i in ticket_price.keys():
    df = pd.DataFrame({'event_name':i[0], 
                       'section_name':i[1],
                       'row_name':i[2],
                       'seat_num':i[3],
                       'event_date':ticket_price[i][0],
                       'Orig_purchase_price':ticket_price[i][1],
                       'add_datetime':ticket_price[i][2],
                       'final_price':ticket_price[i ][3],
                       'sales_channel':ticket_price[i][4],
                       'season_year':ticket_price[i][5],
                       'target':ticket_price[i][6]},
                     index = [0])
    secondary_new = secondary_new.append(df,ignore_index = True)
```
